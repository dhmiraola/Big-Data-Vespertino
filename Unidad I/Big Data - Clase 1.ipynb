## Creacion de variables dummies

```python
# Crear variables dummies
subset['sexo_num'] = pd.get_dummies(subset['sexo'],  drop_first=True)
subset.head()
```

```python
# La variable dummie creada esta en formato bool
subset['sexo_num'].dtypes
```

```python
# Transformar la columna dummies a int
subset['sexo_num'] = pd.get_dummies(subset['sexo'],  drop_first=True).astype(int)
subset.head()
```

```python
# Funcion para clasificar los rangos de edades
def rango_edad(edad):
    if edad <= 18:
        return 'menores_18'
    elif 19 <= edad <= 35:
        return '19-35'
    elif 36 <= edad <= 50:
        return '36-50'
    elif 51 <= edad <= 65:
        return '51-65'
    else:
        return '65+'
```

```python
# Aplicamos la funcion a la columna edad y creamos una carpeta con los nuevos datos 
subset['rango_edad'] = subset['edad'].apply(rango_edad)
subset.head()
```

## Identificacion de Outliers o valores atipicos

```python
# Identificacion de outliers mediante el metodo rango intercuartil
Q1 = data['edad'].quantile(0.25) # 1er cuartil 25%
Q3 = data['edad'].quantile(0.75) # 3er cuartil 75%
IQR = Q3 - Q1

outliers = subset[(subset['edad'] < (Q1 - 1.5 * IQR)) | (subset['edad'] > (Q3 + 1.5 * IQR))]
outliers

# Se puede detectar un outlier en la columna 69902
```

```python
# Identificacion del outliers mediante el metodo de boxplot
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.boxplot(subset['edad'].dropna()) # Usamos dropna() para omitir valores NaN
plt.title('edad')
plt.show()
```
