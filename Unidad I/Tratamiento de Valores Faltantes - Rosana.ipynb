### Tratamiento de Valores Faltantes

```python
# Devuelve la suma de valores perdidos por columna
data.isnull().sum()
```

```python
data.dtypes
```

```python
# creamos una copia del dataset original 
data1 = data.copy()

# Imputar valores perdidos mediante el promedio    
data1["body"] = data1["body"].fillna(data1["body"].mean())
data1["age"] = data1["age"].fillna(data1["age"].mean())
data1["fare"] = data1["fare"].fillna(data1["fare"].mean())

# Corroboramos valores NaN
data1.isnull().sum()
```

```python
# Imputar valores perdidos columnas tipo object

data1['cabin'] = data1['cabin'].fillna('Sin Informacion')
data1['embarked'] = data1['embarked'].fillna('Sin Informacion')
data1['boat'] = data1['boat'].fillna('Sin Informacion')
data1['home.dest'] = data1['home.dest'].fillna('Sin Informacion')

data1.isnull().sum()
```

```python
data2 = data.copy()

# Imputamos los valores NaN mediante la fila de adelante
data2['cabin'] = data2["cabin"].fillna(data["cabin"].ffill())
data2['embarked'] = data2["embarked"].fillna(data["embarked"].ffill())
data2['boat'] = data2["boat"].fillna(data["boat"].ffill())
data2['home.dest'] = data2["home.dest"].fillna(data["home.dest"].ffill())
data2["body"] = data2["body"].fillna(data["body"].ffill())
data2["age"] = data2["age"].fillna(data["age"].ffill())
data2["fare"] = data2["fare"].fillna(data["fare"].ffill())

data2.isnull().sum()
```

```python
data3 = data.copy()

# Imputamos los valores NaN mediante la fila de atras
data3['cabin'] = data3["cabin"].fillna(data["cabin"].bfill())
data3['embarked'] = data3["embarked"].fillna(data["embarked"].bfill())
data3['boat'] = data3["boat"].fillna(data["boat"].bfill())
data3['home.dest'] = data3["home.dest"].fillna(data["home.dest"].bfill())
data3["body"] = data3["body"].fillna(data["body"].bfill())
data3["age"] = data3["age"].fillna(data["age"].bfill())
data3["fare"] = data3["fare"].fillna(data["fare"].bfill())

data3.isnull().sum()
# Quedan valores faltantes, esto debido a que en la columna de atras el dato igualmente era NaN
```

### Borrar datos faltantes

```python
data4 = data.copy()

# Borra las filas que contengan todas las columnas NaN
data4 = data4.dropna(axis=0, how="all") 
data4.isnull().sum()
```

```python
data5 = data.copy()

# Borra las filas que contengan al menos una columna NaN
data5 = data5.dropna(axis=0, how="any") 
data5.head()
```

### Visualizacion de datos Seaborn

```python
import seaborn as sns
# import matplotlib.pyplot as plt # (Esta libreria debes importarla igual)
```

```python
# Histograma de frecuencias
plt.figure(figsize=(8, 4))
sns.histplot(data['age'], kde=True, bins=30)
plt.title('Distribución de Edades en el Titanic')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
# Grafico de violin
plt.figure(figsize=(12, 6))
sns.violinplot(x='clase', y='age', hue='survived', data=data, split=True)
plt.title('Distribución de Edades por Clase y Supervivencia')
plt.xlabel('Clase')
plt.ylabel('Edad')
plt.legend(title='Supervivió', labels=['No', 'Sí'])
plt.show()
```

```python
# Grafico de barras
plt.figure(figsize=(10, 6))
sns.countplot(x='clase', hue='survived', data=data)
plt.title('Supervivencia por Clase de Ticket')
plt.xlabel('Clase')
plt.ylabel('Número de Pasajeros')
plt.legend(title='Sobrevivio', labels=['No', 'Sí'])
plt.show()
```

```python
# Grafico de correlacion
correlacion = data[['age', 'sibsp', 'parch', 'fare']].corr()  # Selecciona solo las columnas numéricas relevantes
plt.figure(figsize=(8, 4))
sns.heatmap(correlacion, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlación entre Características Numéricas')
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='clase', y='fare', hue='survived', data=data)
plt.title('Distribución de la Tarifa por Clase y Supervivencia')
plt.xlabel('Clase')
plt.ylabel('Tarifa')
plt.legend(title='Sobrevivio', labels=['No', 'Sí'])
plt.show()
```

```python
# pairplot
sns.pairplot(data[['age', 'sibsp', 'parch', 'fare', 'survived']], hue='survived')
plt.show()
```

```python
# Grafico de densidad
plt.figure(figsize=(8, 4))
sns.kdeplot(data[data['survived']==1]['age'], label='Sobrevivieron', shade=True)
sns.kdeplot(data[data['survived']==0]['age'], label='No Sobrevivieron', shade=True)
plt.title('Distribución de Edades por Estado de Supervivencia')
plt.xlabel('Edad')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

```python
# Strip plot
plt.figure(figsize=(8, 4))
sns.stripplot(x='survived', y='fare', data=data, jitter=True, alpha=0.5)
plt.title('Distribución de Tarifas por Estado de Supervivencia')
plt.xlabel('Supervivió')
plt.ylabel('Tarifa')
plt.show()
```
