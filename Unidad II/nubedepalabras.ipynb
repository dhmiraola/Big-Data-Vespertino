
Generando una Simple Nube de Palabras
===============

Generaremos una nube de palabras con los poemas del libro "20 Poemas De Amor Y Una Cancion Desesperada" Pablo Neruda


```python
# Si no está instalada la biblioteca wordcloud, con esta línea de código se instala
# !pip install wordcloud
```

```python
import os

from os import path
from wordcloud import WordCloud
import matplotlib.pyplot as plt

import os
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Define la ruta completa al archivo del poema de Neruda
file_path = r'C:\Users\Usuario\Big Data\Text Mining\1 Nube de palabras\neruda.txt'

# Asegúrate de que el archivo existe antes de intentar leerlo
if not os.path.exists(file_path):
    print("El archivo no existe en la ruta especificada.")
else:
    # Leer el archivo con los poemas de Neruda
    text = open(file_path, encoding='utf-8').read()

    # Generar una imagen de nube de palabras
    wordcloud = WordCloud().generate(text)

    # Desplegar la imagen generada
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")

    # Generar una segunda imagen, limitando el tamaño máximo de palabra
    wordcloud = WordCloud(max_font_size=40).generate(text)
    plt.figure()
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```

### Modificando la nube de palabra
Un problema que tiene esta nube de palabras son las palabras comunes en el idioma español (el, la, los, las, de, etc.). Otro problema es que nos gustaría aumentar su tamaño y modificar el color de fondo.

Para más información, revisar la documentación de la biblioteca wordcloud en https://amueller.github.io/word_cloud/auto_examples/index.html

```python
from matplotlib import figure 

# Cargamos las stop words en español
es_stopwords = open("stopwords-es.txt").read().splitlines()

stopwords = set()
stopwords.update(es_stopwords)

# Generate una images de word cloud, esta vez fijamos el color de fondo en blanco
wordcloud = WordCloud(stopwords=stopwords, max_font_size=60, background_color="white").generate(text)

# Desplegamos la imagen usando matplotlib
plt.figure(dpi=150)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
